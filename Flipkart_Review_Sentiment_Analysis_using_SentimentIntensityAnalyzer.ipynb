```python
#cleaning the data
stemmer = nltk.SnowballStemmer('english')

def clean(text):
  text = str(text).lower()
  text = re.sub('\[.*?\]','',text)
  text = re.sub('https?://\S+|WWW\.\S+','',text)
  text = re.sub('<.*?>+','', text)
  text = re.sub('\n','',text)
  text = re.sub('[%s]' % re.escape(string.punctuation),'', text)


  text = [word for word in text.split(' ')]
  text = " ".join(text)
  text = [stemmer.stem(word) for word in text.split(' ')]
  text = " ".join(text)

  return text
```

```python
data['Review'] = data['Review'].apply(clean)
```

```python
#visualize

ratings = data['Rating'].value_counts()
numbers = ratings.index
quantity = ratings.values

import plotly.express as px

figure = px.pie(data, values = quantity, names = numbers, hole=0.5)

figure.show()
```

```python
nltk.download('vader_lexicon')
```

```python
sentiment = SentimentIntensityAnalyzer()
```

```python
data['Postive'] = [sentiment.polarity_scores(i)['pos'] for i in data['Review']]
data['Negative'] = [sentiment.polarity_scores(i)['neg'] for i in data['Review']]
data['Netural'] = [sentiment.polarity_scores(i)['neu'] for i in data['Review']]
```

```python
data = data[["Review","Postive","Negative","Netural"]]
```

```python
data.head(10)
```

```python
x = sum(data['Postive'])
y = sum(data['Negative'])
z = sum(data['Netural'])

def sentiment_score(a,b,c):
  if(a>b) and (a>c):
    print('Postive')
  elif(b>a) and (b>c):
    print('Negative')
  else:
    print('Neutral')

sentiment_score(x,y,z)
```
